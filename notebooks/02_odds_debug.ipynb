```python
import os, requests

api_key = os.getenv("ODDS_API_KEY")
assert api_key, "ODDS_API_KEY not set in this Jupyter session"

url = "https://api.the-odds-api.com/v4/sports/icehockey_nhl/odds"
params = {
    "apiKey": api_key,
    "regions": "us",  # player props are usually US books
    "markets": "player_goal_scorer_anytime",
    "oddsFormat": "decimal",
    "dateFormat": "iso",
}

r = requests.get(url, params=params, timeout=30)
print("Status:", r.status_code)
print(r.text[:500])
```

```python
import os, requests
api_key = os.getenv("ODDS_API_KEY")

url = "https://api.the-odds-api.com/v4/sports/icehockey_nhl/odds"

for market in ["h2h", "totals", "spreads"]:
    r = requests.get(url, params={
        "apiKey": api_key,
        "regions": "eu",
        "markets": market,
        "oddsFormat": "decimal",
        "dateFormat": "iso",
    }, timeout=30)
    print(market, "->", r.status_code)
```
