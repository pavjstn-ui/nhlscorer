```python
python - <<'PY'
import pandas as pd
from pathlib import Path

date = "2025-12-23"
pred_path = Path("data/processed") / f"calibration_snapshot_{date}.csv"
out_path  = Path("data/processed") / f"actual_goals_{date}.csv"
joined_out = Path("data/processed") / f"calibration_joined_{date}.csv"

pred = pd.read_csv(pred_path)
out  = pd.read_csv(out_path)

# Create binary outcome: scored >= 1
out["scored"] = (out["goals"] >= 1).astype(int)

# Keep only what we need from outcomes
out_small = out[["player_norm", "scored", "goals"]].drop_duplicates("player_norm")

joined = pred.merge(out_small, on="player_norm", how="left")

# Players not found in outcomes -> treat as did not score (0)
joined["scored"] = joined["scored"].fillna(0).astype(int)
joined["goals"]  = joined["goals"].fillna(0).astype(int)

joined.to_csv(joined_out, index=False)

print("Saved:", joined_out)
print("Rows:", len(joined))
print("Matched outcomes:", int((joined['scored'].notna()).sum()))
print(joined[["name","team","goal_probability","scored","goals"]].head(10).to_string(index=False))
PY
```
